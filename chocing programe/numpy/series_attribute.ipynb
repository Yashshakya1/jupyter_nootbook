# size ,dtype, name ,is_unique,index,values 
| Attribute    | Works On           | Description                 | Example Output   |
| ------------ | ------------------ | --------------------------- | ---------------- |
| `.name`      | Series             | Name of the Series          | `'Marks'`        |
| `.is_unique` | Series             | Checks for duplicate values | `True / False`   |
| `.index`     | Series / DataFrame | Shows row labels            | `Index([0,1,2])` |
| `.values`    | Series / DataFrame | Returns NumPy array of data | `array([...])`   |


```python
import pandas as pd 
data = pd.Series([100, 200, 300] , index = [18,18,18] , name= 'YR')
print(data)
print(data.is_unique)
print(data.values)
```
