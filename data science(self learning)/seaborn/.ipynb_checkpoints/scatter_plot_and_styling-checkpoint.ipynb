```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

# sns.get_dataset_names()
```

```python
ds = sns.load_dataset('penguins')
top = ds.head()
species = ds['species'].value_counts()
island = ds['island'].value_counts()
sex = ds['sex'].value_counts()
print(top)
print(f'{species}\n {island}\n{sex}')
```

```python
sns.scatterplot(data=ds ,x='flipper_length_mm', y='body_mass_g', hue='island')
plt.show()
```

# The ways of styling themes are as follows:

* white
* dark
* whitegrid
* darkgrid
* ticks


```python
sns.set_style('darkgrid')
sns.scatterplot(data=ds ,x='flipper_length_mm', y='body_mass_g', hue='sex')
plt.show()
```

# The set_context() allows us to override default parameters. 
* This affects things like the size of the labels, lines, and other elements of the plot, but not the overall style. 

# The context are:

* poster
* paper
* notebook
* talk

```python
sns.set_style('dark')
sns.set_context('talk')
sns.scatterplot(data=ds ,x='flipper_length_mm', y='body_mass_g', hue='species')
plt.show()
```

# In Seaborn, despine is a function used to remove the top and right spines (borders) of a plot.
* Default Behavior: Removes the Top and Right borders.

* Remove all 4 borders (Left, Bottom, Right, Top)
* sns.despine(left=True, bottom=True)

*  Keep the right border but remove the others
* sns.despine(right=False, top=True)

* Trim the spines so they stop where the data stops
* (This creates a disconnected look, often seen in scientific papers)
* sns.despine(trim=True)

```python
sns.set_style('ticks')
sns.set_context('paper')
sns.scatterplot(data=ds ,x='flipper_length_mm', y='body_mass_g', hue='species')
sns.despine()
plt.show()
```

```python
sns.set_style('ticks')
sns.set_context('paper')
sns.scatterplot(data=ds ,x='flipper_length_mm', y='body_mass_g', hue='species')
sns.despine(left=True)
plt.show()
```

# a palette determines the color scheme used in your plots.

# The 3 Main Types of Palettes
* Seaborn divides palettes based on the type of data you are visualizing:

* 1. Qualitative (Categorical)
* Best for: Distinct categories that have no order (e.g., Countries, Brands, Gender).

* Goal: To make each category look clearly different.

* Common Names: 'deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind'.

* 2. Sequential
* Best for: Numerical data that goes from Low to High (e.g., Age, Price, Density).

* Goal: To show intensity using lighter to darker shades of a single color (or smooth blends).

* Common Names: 'Blues', 'Greens', 'Reds', 'rocket', 'mako'.

* 3. Diverging
* Best for: Data with a meaningful "center" point (e.g., Temperature diverging from 0, Profit vs. Loss, Agree vs. Disagree).

* Goal: Two contrasting colors (like Red and Blue) meet at a neutral middle (usually white or gray).

* Common Names: 'vlag', 'coolwarm', 'RdBu' (Red-Blue), 'Spectral'.

```python
sns.set_style('ticks')
sns.set_context('notebook')
sns.scatterplot(data=ds ,x='flipper_length_mm', y='body_mass_g', hue='species',palette='rocket')
plt.show()
```
