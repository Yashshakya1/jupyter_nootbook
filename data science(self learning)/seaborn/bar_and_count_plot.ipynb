# bar plot 
* A Barplot (sns.barplot) in Seaborn is used to visualize the average (mean) value of a numerical variable for different categories.
* Crucial Distinction: Unlike a standard Histogram or Countplot (which shows how many items exist), a Barplot calculates a statistic (like the average price, total sales, etc.) for each group.

# Customization: Changing the Calculation (estimator)
* If you don't want the Mean, you can change it to calculate the Sum, Median, or Standard Deviation using the estimator parameter.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
ds = sns.load_dataset('penguins')
top = ds.head()
species = ds['species'].value_counts()
island = ds['island'].value_counts()
sex = ds['sex'].value_counts()
print(top)
print(f'{species}\n {island}\n{sex}')
```

```python
sns.barplot(data=ds,x='species',y='body_mass_g')
```

```python
sns.set_style('dark')
sns.barplot(data=ds,x='species',y='body_mass_g',hue='sex',palette=['green','red'])
```

```python
sns.barplot(data=ds,x='species',y='body_mass_g',hue='sex',estimator=np.var ,palette=['green','purple'])
```

```python
sns.barplot(data=ds,x='species',y='body_mass_g',hue='sex',estimator=np.median ,palette=['green','purple'])
```

# Count plot
* this plot showing how many no of counts in a particular columns using count plot 

```python
sns.set_style('darkgrid')
sns.set_context('talk')
sns.countplot(data=ds,x='island',color='green')
```

```python
sns.set_style('darkgrid')
sns.set_context('talk')
sns.countplot(data=ds,x='island',hue = 'sex',palette='Set1')
```
