# stripplot

 * A Stripplot is essentially a scatter plot designed for categorical data.
# Why use it?
* Standard scatter plots fail with categorical data because the points overlap perfectly, looking like a single straight line. 
*A stripplot solves this by adding "jitter" (random noise) to spread the points out horizontally so you can see them individually.

#  jitter (The Magic Ingredient)
* By default, jitter=True. If you turn it off (jitter=False), 
* all dots will line up in a straight vertical line, making it hard to see how many dots are there.

# dodge
* dodge=True: Splits the strips so the hue categories don't overlap (similar to a grouped bar chart).

```python
# style='sex' showing the error i don't know why 
```

```python
sns.stripplot(data=ds,x='species',y= 'body_mass_g',hue = 'island',alpha = 0.5)
```

```python
sns.stripplot(data=ds,x='species',y= 'body_mass_g',hue = 'island',alpha = 0.5,dodge=True)
```

```python
sns.stripplot(data=ds,x='species',y= 'body_mass_g',hue = 'island',alpha = 0.5,jitter=0.3,dodge=True)
```

# swarmplot
* A Swarmplot is very similar to a Stripplot, but with one major difference: points are not allowed to overlap.

Feature	Stripplot	Swarmplot						
Logic	Adds random noise (jitter) to spread points.	Mathematically arranges points to avoid touching.						
Appearance	Messy, points can overlap.	Clean, structured, shows distribution shape clearly.						
Dataset Size	Good for Large Data. (It can handle thousands of points).	Bad for Large Data. If there are too many points, they will run off the side of the graph, and Seaborn will throw a warning.						

```python
sns.swarmplot(data= ds , x='species',y= 'body_mass_g',hue = 'island')
sns.set_context('talk')
sns.despine()
```
