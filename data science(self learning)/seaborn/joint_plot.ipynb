```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
ds = sns.load_dataset('penguins')
top = ds.head()
species = ds['species'].value_counts()
island = ds['island'].value_counts()
sex = ds['sex'].value_counts()
print(top)
print(f'{species}\n {island}\n{sex}')
```

# A Jointplot (sns.jointplot) is a Seaborn function that combines two plots into one:

* Main Plot (Center): Shows the relationship between two variables (X and Y), usually as a scatter plot.

* Marginal Plots (Sides): Shows the distribution of each variable individually (histograms or density curves) on the top and right axes.

# kind=,Visual,Best For...
* 'scatter'  ,(Default) Dots.,Standard correlation check.
* 'reg'      ,Scatter + Regression Line.,Seeing linear trends and model fit.
* 'hex'      ,Hexagons (Honeycomb).,Large datasets. It groups dots to avoid overlapping mess.
* 'kde'      ,Contour lines (Topographic map).,"Finding clusters or ""hotspots"" in density."

```python
sns.set_style('dark')
sns.jointplot(data=ds,x='body_mass_g',y= 'flipper_length_mm')
```

```python
sns.jointplot(data=ds,x='body_mass_g',y= 'flipper_length_mm',hue='sex',palette='mako')
```

```python
sns.jointplot(data=ds,x='body_mass_g',y= 'flipper_length_mm',kind='reg',color='green')
```

```python
sns.jointplot(data=ds,x='body_mass_g',y= 'flipper_length_mm',hue='sex',kind='kde')
```
