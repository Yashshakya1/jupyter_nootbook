```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
ds = sns.load_dataset('penguins')
top = ds.head()
species = ds['species'].value_counts()
island = ds['island'].value_counts()
sex = ds['sex'].value_counts()
print(top)
print(f'{species}\n {island}\n{sex}')
```

```python
sns.histplot(data=ds,x ='body_mass_g')
```

```python
sns.histplot(data=ds,x ='body_mass_g', hue='sex')
```

```python
sns.histplot(data=ds,x ='body_mass_g', hue='sex',multiple='stack')
```

# Regression plot

```python
sns.regplot(data=ds,x = 'body_mass_g',y ='flipper_length_mm',color='black')  # scatter=False [remove scatter points]
```

# line plot 
* visulition of line plot for the numerical columns in the continous manner 

```python
sns.set_style('darkgrid')
sns.lineplot(data=ds,x='body_mass_g',y='flipper_length_mm')  # palette is not used if there is no hue in your plot  
```

```python
sns.set_style('whitegrid')
sns.lineplot(data=ds,x='body_mass_g',y='flipper_length_mm',hue='sex',style='sex',palette='Set1')
```
