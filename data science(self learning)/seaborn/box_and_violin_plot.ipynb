```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
ds = sns.load_dataset('penguins')
top = ds.head()
species = ds['species'].value_counts()
island = ds['island'].value_counts()
sex = ds['sex'].value_counts()
print(top)
print(f'{species}\n {island}\n{sex}')
```

```python
sns.boxplot(data=ds,x= 'species',y='body_mass_g')
```

```python
sns.boxplot(data=ds,x= 'species',y='body_mass_g',hue='sex')
```

# violinplot

```python
sns.violinplot(data=ds,x= 'species',y='body_mass_g',hue='sex')
```

```python
sns.violinplot(data=ds,x= 'species',y='body_mass_g',hue='sex',splite=True)
```
