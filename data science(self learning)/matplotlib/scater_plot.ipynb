```python
import matplotlib.pyplot as plt 
import numpy as np 
hours = np.random.randint(0,16,10)
scores =np.random.randint(1,99,10)
plt.scatter(scores,hours, color = 'red', label = "student data", marker = "D")
plt.legend(loc = "lower right")
plt.xlabel("scores")
plt.ylabel("study hours ")
plt.title("student scores anylises")
plt.grid(True,color = 'black', axis= 'both',ls = ':',lw = 1,alpha = 0.5)
plt.show()
```

# two scatter data points in the one graph

```python
import matplotlib.pyplot as plt 
import numpy as np 
plt.scatter([1,2,3,4],[50,60,30,95] ,color = 'red', label = "class A", marker = "o")
plt.scatter([1,2,3,4],[40,50,70,65] ,color = 'green', label = "class B", marker = "D")
plt.legend(loc = "upper left")
plt.xlabel("scores")
plt.ylabel("study hours ")
plt.title("student scores anylises")
plt.grid(True)
plt.show()
```
