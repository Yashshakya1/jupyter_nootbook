```python
import matplotlib.pyplot as plt 

x = [1,2,3,4,5,6]
y = [10,25,30,45,50,70]
plt.subplot(1,2,1)
plt.plot(x,y, c = "green")
plt.title("line charts")

plt.subplot(1,2,2)
plt.bar(x,y, color = "green")
plt.title("bar chart")
plt.suptitle("this is the title baby")
plt.tight_layout()
plt.show()
```
