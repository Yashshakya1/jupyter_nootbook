```python
import matplotlib.pyplot as plt 

x = [1,2,3,4,5,6]
y = [10,25,30,45,50,70]
plt.subplot(1,2,1)
plt.plot(x,y, c = "green")
plt.title("line charts")

plt.subplot(1,2,2)
plt.bar(x,y, color = "green")
plt.title("bar chart")
plt.suptitle("this is the title baby")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt 

x = [1,2,3,4,5,6]
y = [10,25,30,45,50,70]
fig, ax = plt.subplots(1,2, figsize = (8,5))
# ax[0].plot(x,y,'^', c = 'red')
# ax[0].grid(True)
ax[0].plot(x,y,c= "green")
ax[0].set_title("line chart bro ")

ax[1].hist(x, bins = 4, color = "purple"  , ec = "black")
ax[1].set_title("histogram  chart ")
fig.suptitle("camparision between the line and bar chart")
plt.tight_layout()
plt.show()
```
