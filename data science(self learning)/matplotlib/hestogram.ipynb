```python
# graph that shows how data is distributed — it displays the frequency (count) 
# of data points that fall within specific ranges (called bins).
import matplotlib.pyplot as plt 
import numpy as np 
scros = np.random.randint(1,100,200)
plt.hist(scros,bins = 5, color = "purple" , ec = "black",label="data")
print(scros)
plt.xlabel("scores")
plt.ylabel("student")
plt.title("student marks between the 50")
plt.legend()
plt.show()
```
