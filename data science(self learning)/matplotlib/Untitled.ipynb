```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
ds = pd.read_csv('cricket.csv')
ds
```

```python
x = ds['Year']
y = ds['Virat']
plt.plot(x,y,ls =':',marker = 'o',c='#367588',label = 'year')
plt.plot(x,ds['Rohit'],c ='green',label = 'rohit',linewidth = '5' )
plt.title('goat player runs comparasion')
plt.xlabel('year')
plt.ylabel('score')
plt.xlim(2010,2016)
plt.ylim()
plt.legend(loc = 'lower right') # by defulat loc is best 
plt.tight_layout()
plt.show()
```
