```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

* ['tips', 'iris', 'titanic', 'flights', 'penguins', 'planets', ...]

```python
ds = sns.load_dataset('tips')
ds
```

```python
plt.scatter(ds['total_bill'], ds['tip'])
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title('corelation between the total_bill and tip')
# plt.grid(True,c = 'black') 
plt.tight_layout()
plt.show()
```

```python
plt.scatter(ds['day'],ds['total_bill'])
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title('corelation between the total_bill and tip')
# plt.grid(True,c = 'black') 
plt.tight_layout()
plt.show()
```
