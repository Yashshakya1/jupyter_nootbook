```python
import numpy as np
import matplotlib.pyplot as plt

year = [2010, 2012, 2014]
score = np.array([[98,60,72],
                  [100,85,89],
                  [99,98,91]])

x = np.arange(len(year))   # [0,1,2]
width = 0.25

plt.bar(x - width, score[:,0], width, label='Score 1')
plt.bar(x,         score[:,1], width, label='Score 2')
plt.bar(x + width, score[:,2], width, label='Score 3')

plt.xticks(x, year)
plt.xlabel('Year')
plt.ylabel('Score')
plt.title('Scores Comparison Using Bar Chart')
plt.legend()
plt.grid(axis='y',c='black')

plt.show()
```

# another method to create a same multi bar plot

```python
for i in range(3):
    plt.bar(x + (i-1)*width, score[:,i], width)

for i in range(3):
    for j in range(len(year)):
        plt.text(x[j] + (i-1)*width, score[j][i]+1,
                 score[j][i], ha='center')

plt.xticks(x, year)
plt.xlabel('Year')
plt.ylabel('Score')
plt.title('Scores with Values')
plt.show()
```

```python
import numpy as np
labels = ['2019', '2020', '2021']
a = [10, 15, 20]
b = [12, 18, 25]

x = np.arange(len(labels))
total = np.array(a) + np.array(b)
plt.bar(x, a/total*100, label='A')
plt.bar(x, b/total*100, bottom=a/total*100, label='B')
plt.legend()
plt.show()
```

# stakeing the bar charts with array

```python
import numpy as np
import matplotlib.pyplot as plt

year = [2010, 2012, 2014]
score = np.array([
    [98, 60, 72],
    [100, 85, 89],
    [99, 98, 91]
])

x = np.arange(len(year))

plt.bar(x, score[0], label='Student A')
plt.bar(x, score[1], bottom=score[0], label='Student B')
plt.bar(x, score[2], bottom=score[0] + score[1], label='Student C')

plt.xticks(x, year)
plt.xlabel('Year')
plt.ylabel('Score')
plt.title('Stacked Bar Graph')
plt.legend()

plt.show()
```

```python
import numpy as np
countries = ['usa','india','china','japan']
gf = [2,15,49,56]
bf = [12,56,89,23]
x = [7,48,59,26]
w = 0.25
h = np.add(gf,bf)
plt.bar(countries,gf,w, label = 'gf',ec = 'black')
plt.bar(countries,bf,w,bottom=gf,color = 'green')
plt.bar(countries,x,w,bottom=h,color = 'pink')
plt.legend(title = 'relationship')
plt.show()
```
