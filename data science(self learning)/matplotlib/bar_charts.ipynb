```python
import matplotlib.pyplot as plt 
product = ['realme','poco','moto','apple','vivo']
sales = [2000,500,1000,10000,8000]
plt.bar(product,sales, color = "green" , label = 'sales 2025')
plt.legend()
plt.xlabel('product --')
plt.ylabel('sales')
plt.title("mobile sales 2025 in india ")
plt.show()
```

```python
import matplotlib.pyplot as plt 
x = ["pen","pencil","scale","glow"]
y = [60,100,30,46]
plt.barh(x,y,label = "sales 2025", color = "red")
plt.legend()
plt.title("stationary item sales ")
plt.xlabel("items")
plt.ylabel("sales")
```

# secand method by using subplot method 

```python
import matplotlib.pyplot as plt

names = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

fig, ax = plt.subplots()
bars = ax.bar(names, values)

ax.bar_label(bars)   #  show values on bars

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

year = [2010, 2012, 2014]
score = np.array([[98,60,72],
                  [100,85,89],
                  [99,98,91]])

x = np.arange(len(year))   # [0,1,2]
width = 0.25

plt.bar(x - width, score[:,0], width, label='Score 1')
plt.bar(x,         score[:,1], width, label='Score 2')
plt.bar(x + width, score[:,2], width, label='Score 3')

plt.xticks(x, year)
plt.xlabel('Year')
plt.ylabel('Score')
plt.title('Scores Comparison Using Bar Chart')
plt.legend()
plt.grid(axis='y',c='black')

plt.show()
```

# another method to create a same multi bar plot

```python
for i in range(3):
    plt.bar(x + (i-1)*width, score[:,i], width)

for i in range(3):
    for j in range(len(year)):
        plt.text(x[j] + (i-1)*width, score[j][i]+1,
                 score[j][i], ha='center')

plt.xticks(x, year)
plt.xlabel('Year')
plt.ylabel('Score')
plt.title('Scores with Values')
plt.show()
```
