```python
import matplotlib.pyplot as plt 
product = ['realme','poco','moto','apple','vivo']
sales = [2000,500,1000,10000,8000]
plt.bar(product,sales, color = "green" , label = 'sales 2025')
plt.legend()
plt.xlabel('product --')
plt.ylabel('sales')
plt.title("mobile sales 2025 in india ")
plt.show()
```

```python
import matplotlib.pyplot as plt 
x = ["pen","pencil","scale","glow"]
y = [60,100,30,46]
plt.barh(x,y,label = "sales 2025", color = "red")
plt.legend()
plt.title("stationary item sales ")
plt.xlabel("items")
plt.ylabel("sales")
```
