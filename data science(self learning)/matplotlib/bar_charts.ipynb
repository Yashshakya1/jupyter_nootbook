```python
import matplotlib.pyplot as plt 
product = ['realme','poco','moto','apple','vivo']
sales = [2000,500,1000,10000,8000]
plt.bar(product,sales, color = "green" , label = 'sales 2025')
plt.legend()
plt.xlabel('product --')
plt.ylabel('sales')
plt.title("mobile sales 2025 in india ")
plt.show()
```

```python
import matplotlib.pyplot as plt 
x = ["pen","pencil","scale","glow"]
y = [60,100,30,46]
plt.barh(x,y,label = "sales 2025", color = "red")
plt.legend()
plt.title("stationary item sales ")
plt.xlabel("items")
plt.ylabel("sales")
```

```python
plt.bar(x,y,width=0.3) # The default width value is 0.8
plt.show()
```

```python
plt.barh(x,y,height=0.1)# The default height value is 0.8
plt.show()
```

```python
plt.bar(x,y)
for i in range(len(x)):
    plt.text(x[i],y[i],y[i] , ha ='center',va = 'center')
plt.show()
```

# secand method by using subplot method 

```python
import matplotlib.pyplot as plt

names = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

fig, ax = plt.subplots()
bars = ax.bar(names, values)

ax.bar_label(bars)   #  show values on bars

plt.show()
```
