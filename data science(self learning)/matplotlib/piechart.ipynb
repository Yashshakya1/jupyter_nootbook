# Explode
* Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

* The explode parameter, if specified, and not None, must be an array with one value for each wedge.

* Each value represents how far from the center each wedge is displayed:

# Shadow
* Add a shadow to the pie chart by setting the shadows parameter to True:

```python
python_library = ['Numpy','Pandas','Matplotlib','Seaborn','Ploty']
most_used = [50,80,60,55,35]
plt.pie(most_used,labels=python_library,autopct='%1.f%%',colors= ['red','purple','green','pink','yellow'],startangle=90,explode=[0,0,0.2,0,0]
       ,shadow=True)
plt.legend(title = 'most used library in ml')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Create a range of values (t)
t = np.linspace(0, 2 * np.pi, 1000)

# 2. The Heart Equation
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# 3. Plot it
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='red', linewidth=3)
plt.fill(x, y, color='pink') # Optional: Fill it with color

plt.title("Heart Plot for special person t name ")
plt.axis('equal') # Keeps the aspect ratio correct
plt.axis('off')   # Hides the axes for a cleaner look
plt.show()
```
