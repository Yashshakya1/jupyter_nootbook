```python
# plt.plot(x,y, color = "blue" , linestyle, linewidth = value ,marker = 'D', label = 'name' )
import matplotlib.pyplot as plt 
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([100,200,300,400,500])
plt.plot(x,y, color = "red", linestyle = ':' , linewidth = 2 , marker = 'D', label = 'khala ga free fire ')
plt.xlabel("num of product")
plt.ylabel("quantity")
plt.title("product stocks in shop", loc = "right")
plt.legend(loc = "lower right",fontsize = 10)
plt.grid(color = "green", linestyle = ":", linewidth = 1)
plt.xlim(1,4)
plt.xticks([1,2,3,4,5], ['p1','p2','p3','p4','p5'])
plt.show()
```
