# 1.# Create a 2D array with 1 on border and 0 on inside.

```python
import numpy as np 
arr = np.ones(100).reshape(10,10)
arr[1:-1, 1:-1] = 0
print(arr)
```

# 2 .Find indices of Non-zero elements from A[2,3,0,1,0,9].

```python
A = np.array([2,3,0,1,0,9])
index = np.nonzero(A) # fast and simple 
ind = np.where(A != 0) # fast but best use for conditional stuation 
inde = np.arange(len(A))[A != 0] # slightly longer but if you went output in array then this is upmm  
print(inde)
print(index)
print(ind)
```

# 7 Sorting a NumPy array based on a specific column. Ex: Array = np.array([[34, 43, 73], [82, 22, 12], [53, 94, 66]] Note:(Sort records according 2nd column)

```python
ar = np.array([[34,43,73] , 
               [82,22,12] , 
               [53,94,66]])
sort = ar[ar[:,1].sort() ]
print(sort)
```
