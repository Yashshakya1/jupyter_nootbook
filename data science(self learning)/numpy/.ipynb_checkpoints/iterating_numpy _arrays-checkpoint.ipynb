# nditer() Ndenumerate()

```python
import numpy as np
```

```python
arr = np.array([0,12,4,5,9])
for i in arr:
    print(i)
```

```python
arr = np.array([[1,2,3,4,5,6],[6,5,4,1,2,3]])
for i in arr:
    print(i)
```

```python
arr = np.array([[1,2,3,4,5,6],[6,5,4,1,2,3]])
for i in arr:
    for j in i:
        print(j) 
```

# nditer() method is used for iterating the numpy array elements 

```python
arr = np.array([[1,2,3,4,5,6],[6,5,4,1,2,3]])
for i in np.nditer(arr):
    print(i)
```

```python
arr3 = np.array([[[1,2,3,4,5],[5,4,1,2,3]] , [[7,8,9,4,5],[5,4,9,8,7]]])
print(arr3,'\n dimension :-',arr3.ndim)
for i in arr3:
    print(i)
```

```python
arr3 = np.array([[[1,2,3,4,5],[5,4,1,2,3]] , [[7,8,9,4,5],[5,4,9,8,7]]])
print(arr3,'\n dimension :-',arr3.ndim)
for i in arr3:
    for j in i:
        for k in j:
            print(k)
```

```python
arr3 = np.array([[[1,2,3,4,5],[5,4,1,2,3]] , [[7,8,9,4,5],[5,4,9,8,7]]])
print(arr3,'\n dimension :-',arr3.ndim)
for i in np.nditer(arr3):
    print(i)
```

```python
# np.ndenumerate(array_value)
# return index as a tuple (index,value/data)

arr_ind = np.arange(8).reshape(4,2)
for i,d in np.ndenumerate(arr_ind): 
    print(i,d)
```
