# endcoding 

 * encoding means converting the data from one format into onother format is known as encoding 
 * why need this because ml don't read the string they only know the numbers like 0 or 1
 * types of encoding:-
 * 1. label encoding
 * 2. one hot encodign
 * 3. Ordinal Encoding (Used when the categories have a natural order.) like :-
 * | Education   | Encoded |
| ----------- | ------- |
| High School | 1       |
| Bachelor’s  | 2       |
| Master’s    | 3       |
| PhD         | 4       |

* 4️⃣ Frequency / Count Encoding (Replace each category with how often it appears.)
| City   | Frequency |
| ------ | --------- |
| Delhi  | 3         |
| Mumbai | 2         |
| Delhi  | 3         |
| Pune   | 1         |

* 5️⃣ Binary / Hash / Target Encoding (Used for large datasets or high-cardinality features (e.g., 1000+ categories).)

   # this is summary tabel
* | Encoding Type         | Best for             | Description                          |
| --------------------- | -------------------- | ------------------------------------ |
| **Label Encoding**    | Simple categories    | Assigns numeric labels (0, 1, 2, …)  |
| **One-Hot Encoding**  | Unordered categories | Creates binary columns               |
| **Ordinal Encoding**  | Ordered categories   | Keeps natural rank (low → high)      |
| **Count / Frequency** | Large categories     | Uses counts of occurrences           |
| **Target Encoding**   | Advanced ML          | Uses relationship to target variable |

* 


```python
# label encoding 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

li = ['Red','Blue','Green','Red']
encoding = le.fit_transform(li)
print(encoding)
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd 
ds = pd.DataFrame({
           'name': ['yash','rani','rani kamlapati','suraj','roshni'],
            'gender': ['M','F','F','M','F'],
          'passed': ['yes','yes','no','yes','no']
})
le = LabelEncoder()
ds_label = ds.copy()

ds_label['Gender_endcoded'] = le.fit_transform(ds_label['gender'])
ds_label['Passed_endcoded'] = le.fit_transform(ds_label['passed'])
print(ds_label)
```

# one hot coding
 * pd.get_dummies() is a pandas function used to perform One-Hot Encoding automatically.

```python
# one hot coding
import pandas as pd 
df = pd.DataFrame({
                'colours':  ['Red','Blue','Green','Red']
})
encodeing = pd.get_dummies(df,columns=['colours'] )  #,dtype=int
print(encodeing)
```
