# logistic regression
* used for the label type prediction like spam yes or not and always outcomes between 0 and 1 

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LogisticRegression
# data
x = np.arange(1,15).reshape(-1,1)
y = np.random.randint(0,2,14)


# model 
model = LogisticRegression()
model = model.fit(x,y)
hours = float(input("how many hours you study :-"))
print(y)
# prediction
predict = model.predict([[hours]])
if predict == 1:
    print(f"✅ Based on your study hours ({hours}), you are **PASS**!")
else:
    print(f"❌ Based on your study hours ({hours}), you are **FAIL**!")
```
