```python
import numpy as np 
from scipy import stats
a =np.random.randint(0,100,10)
b =np.random.randint(0,100,10)
t_test,p_value = stats.ttest_ind(a,b,equal_var = False)
```

```python
print(f'{a}\n{b}')
```

```python
print(f'{t_test}\n{p_value}')
```

```python
alpa = 0.05
if p_value<alpa:
    print('i will reject your null hypothises test')
else:
    print('i will accept your null hypothises test ')
```
