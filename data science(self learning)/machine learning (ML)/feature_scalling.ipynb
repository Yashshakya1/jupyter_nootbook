# Feature scalling
* Feature Scaling means normalizing or standardizing the range of your numerical data features
  so that all features are on a similar scale.
* It makes sure that big numbers don‚Äôt dominate small ones in your dataset.
* ex -
* | Height (cm) | Weight (kg) |
| ----------- | ----------- |
| 170         | 60          |
| 180         | 80          |
| 160         | 50          |

* if i add income
* | Height | Weight | Income  |
| ------ | ------ | ------- |
| 170    | 60     | 800000  |
| 180    | 80     | 1000000 |
| 160    | 50     | 600000  |

* üü¢ 1Ô∏è‚É£ Normalization (Min‚ÄìMax Scaling)
* Scales all values to a fixed range, usually 0 to 1.

* üîµ 2Ô∏è‚É£ Standardization (Z-Score Scaling)
* Centers data around mean = 0 and standard deviation = 1. 

* ‚ö™ 3Ô∏è‚É£ Robust Scaling (Handles Outliers)
* Scales using median and interquartile range (IQR) ‚Äî less affected by outliers.

| Method                        | Formula                 | Range    | Sensitive to Outliers? | Common Use                      |
| ----------------------------- | ----------------------- | -------- | ---------------------- | ------------------------------- |
| **Min‚ÄìMax**                   | (X - min) / (max - min) | 0 to 1   | ‚úÖ Yes                  | Neural Nets, KNN                |
| **Standardization (Z-Score)** | (X - mean) / std        | Around 0 | ‚ö†Ô∏è Less                | Linear/Logistic Regression, SVM |
| **Robust Scaling**            | (X - median) / IQR      | Around 0 | ‚ùå No                   | Data with outliers              |


# StandardScaler() 

```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split

data = {
    'study_hours': [1,5,2,4,4,6],
    'test_score' : [60,50,8,8,80,15]
}
df = pd.DataFrame(data)
scaler = StandardScaler()
standard_scaled = scaler.fit_transform(df)
print(pd.DataFrame(standard_scaled, columns=['study_hours','test_score']))
```

# MaxMinScaler

```python
mn_scaler = MinMaxScaler()
mn_scaled = mn_scaler.fit_transform(df)
print(pd.DataFrame(mn_scaled, columns=['study_hours','test_score']))
```

# train and test your model (splitting)

```python
from sklearn.model_selection import train_test_split
x = df[['study_hours']]
y = df[['test_score']]
x_train,x_test,y_train,y_test = train_test_split(x,y , test_size=0.2 , random_state=42)
print('train data')
print(x_train,'\ntest data',x_test)

print(y_train,y_test)
```
