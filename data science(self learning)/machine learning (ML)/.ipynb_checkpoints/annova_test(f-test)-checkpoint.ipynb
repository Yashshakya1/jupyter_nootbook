```python
import seaborn as sns
import pandas as pd 
from scipy.stats import f_oneway 
df = sns.load_dataset('titanic')
df = df[['age','pclass']].dropna()
top_5_values = df.head()
print(top_5_values)
df['pclass'].unique()
```

```python
# create the class 
class1 = df[df['pclass'] == 1]['age']
class2 = df[df['pclass'] == 2]['age']
class3 = df[df['pclass'] == 3]['age']
```

```python
f_test,p_values = f_oneway(class1,class2,class3)
print(f_test)
print(p_values)
```

```python
alpa = 0.05
if p_values<alpa:
    print('h0 is rejected ')
else:
    print('h0 is acepted ')
```
