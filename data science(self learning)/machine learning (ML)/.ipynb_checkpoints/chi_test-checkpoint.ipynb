```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
from scipy.stats import chi2_contingency
ds = sns.load_dataset('titanic')
ds.head()
```

```python
ds.columns
```

# crosstab() fuction or method 
* It shows how many times combinations of values occur.

```python
contangency_table = pd.crosstab(ds['sex'],ds['survived'])
print(contangency_table)
```

```python
chi2,p_value,dof,expected = chi2_contingency(contangency_table) # chi2,p_value,
print(expected)
```

```python
print(f'{chi2}\n{p_value}\n{dof}')
```

```python
alpa = 0.05
if p_value<alpa:
    print('h0 is rejected ')
else:
    print('h0 is acepted ')
```
