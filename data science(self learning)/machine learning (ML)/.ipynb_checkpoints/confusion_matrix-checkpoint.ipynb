# What is a confusion matrix?
* It is a table that is used in classification problems to assess where errors in the model were made.

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import metrics
# data
actual = np.random.binomial(1,.9, size= 1000)
prediction = np.random.binomial(1,.9, size = 1000)

confusion_matrix = metrics.confusion_matrix(actual,prediction)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix, display_labels= [0,1])
cm_display.plot()
plt.tight_layout()
plt.show()
```

# True Negative (Top-Left Quadrant)
# False Positive (Top-Right Quadrant)
# False Negative (Bottom-Left Quadrant)
# True Positive (Bottom-Right Quadrant)

* How many predictions were correct (TP, TN)
* How many were wrong (FP, FN)
* The confusion matrix helps evaluate the performance of a logistic regression model

| Metric        | Formula                                         | Meaning                                           |
| ------------- | ----------------------------------------------- | ------------------------------------------------- |
| **Accuracy**  | (TP + TN) / Total                               | Overall correctness                               |
| **Precision** | TP / (TP + FP)                                  | When model says "Positive", how often is it right |
| **Recall**    | TP / (TP + FN)                                  | How many actual positives did it find             |
| **F1 Score**  | 2 × (Precision × Recall) / (Precision + Recall) | Balance between Precision & Recall                |
