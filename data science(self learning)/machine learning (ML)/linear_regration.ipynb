```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
# data
x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([10,30,50,70,90])

# model
model = LinearRegression()
model.fit(x,y)
# Prediction
y_pred = model.predict(x)
# plotting
plt.scatter(x,y, color = 'red', label = 'Actual data')
plt.plot(x,y_pred ,color = 'green' ,label ='fit line')
plt.title('example of linear regression')
plt.xlabel('Study hours')
plt.ylabel('Marks')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Training data
x_axis = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y_axis = np.array([10, 30, 50, 70, 90])

# Create model and fit
model = LinearRegression()
model.fit(x_axis, y_axis)

# Take user input
hours = float(input('Enter your study hours: '))

# Predict
pred = model.predict([[hours]])

# Output
print(f'Study hours: {hours} â†’ Predicted score: {pred[0]:.2f}')
```
