# ğŸ§  What is KNN (K-Nearest Neighbors)?

* KNN stands for K-Nearest Neighbors.
* It is a supervised learning algorithm that predicts the output for a new data point based on the majority of its nearest neighbors in the training data.

# example

| Study Hours | Score | Result |
| ----------- | ----- | ------ |
| 1           | 20    | Fail   |
| 2           | 30    | Fail   |
| 3           | 50    | Pass   |
| 4           | 70    | Pass   |

* Now if a new student studied 2.5 hours, KNN looks at the K nearest points (say K=3) around 2.5 and checks their labels.
* ğŸ‘‰ If most of them are â€œPassâ€, KNN predicts Pass.
* ğŸ‘‰ If most are â€œFailâ€, KNN predicts Fail.

```python
import numpy as np 
from sklearn.neighbors import KNeighborsClassifier
# data
x = np.array([[180,7],[200,7.5],[250,8],[300,8.5],[350,9]])
y = np.random.randint(0,2,5)

# model 
model = KNeighborsClassifier(n_neighbors=3)
model = model.fit(x,y)

weight = float(input("enter your weight :-"))
size = float(input("enter your size :-"))
print(y)
predection = model.predict([[weight,size]])[0]
if predection == 1:
    print('orange')
else:
    print('apple')
```
