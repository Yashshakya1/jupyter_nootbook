```python
import pandas as pd

# Load dataset
df = pd.read_csv("ipl data set.CSV")

# ---- Batting Analysis ----
batting_stats = df.groupby("Team").agg({
    "Runs": "sum",
    "StrikeRate": "mean",
    "100s": "sum",
    "50s": "sum"
}).reset_index().sort_values("Runs", ascending=False)

# ---- Bowling Analysis ----
bowling_stats = df.groupby("Team.1").agg({
    "Wickets": "sum",
    "EconomyRate": "mean"
}).reset_index().sort_values("Wickets", ascending=False)

# ---- Top Individual Performers ----
top_batsmen = df.sort_values("Runs", ascending=False).head(5)[
    ["Season","Player(Batting)","Team","Runs","StrikeRate","100s","50s"]
]

top_bowlers = df.sort_values("Wickets", ascending=False).head(5)[
    ["Season","Player(Bowling)","Team.1","Wickets","BestFigures","EconomyRate"]
]

print("=== Batting Stats by Team ===")
print(batting_stats)

print("\n=== Bowling Stats by Team ===")
print(bowling_stats)

print("\n=== Top 5 Batsmen ===")
print(top_batsmen)

print("\n=== Top 5 Bowlers ===")
print(top_bowlers)
```
