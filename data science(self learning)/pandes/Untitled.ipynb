# Series do not support -1 indexing 

```python
hasnans return true or false in whole data 
```

```python
import pandas as pd

s = pd.Series([10, None, 30])
print(s.hasnans)
```

```python
import pandas as pd

s = pd.Series([10, None, 30, float('nan')])
print(s.isnull())
```

# Filter non-missing values:

| Function / Attribute       | Works On         | Meaning                    |
| -------------------------- | ---------------- | -------------------------- |
| `s.hasnans`                | Series           | Any NaN exists             |
| `s.isnull()`               | Series/DataFrame | Shows True/False per value |
| `df.isnull().values.any()` | DataFrame        | DataFrame contains NaN     |
