```python
import pandas as pd 
ds = pd.read_excel('weather_data.xlsx')
ds
```

```python
groupby return two par key and values that was c and value in the value 
```

```python
ct = ds.groupby(['city'])
for c,value in ct:
    print(c)
    print(value)
```

# get a column in spacific position with the help of get_group

```python
ct.get_group('bhopal')
```

```python
ct.get_group('bhopal').apply(max)
```

```python
def cybrom(v):
    return v > 18
ct['windspeed'].apply(cybrom)
```

```python
ct.get_group('bhopal').filter(lambda x: ct > 18)
```
