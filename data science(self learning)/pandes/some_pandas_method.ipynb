# between() fuction 
The between() function in Pandas is a method available for Series objects, used to check if values within the Series fall within a specified range. It returns a boolean Series indicating whether each element in the original Series is within the given lower and upper boundaries.

* Series.between(left, right, inclusive='both')

```python
import pandas as pd 
import numpy as np 
data = pd.Series(np.arange(1,11))
data_fillterd = data[data.between(5,10)]
print(data.between(5,10))
print(data_fillterd)
```

```python
# inclusive = 'neither'  print the value between 5 to 7 that was 6 in index 5
data_fillter = data[data.between(5,7, inclusive='neither')]
print(data_fillter)
```

# drop_duplicates 
* DataFrame.drop_duplicates(subset=None, *, keep='first', inplace=False, ignore_index=False)
* Return DataFrame with duplicate rows removed.

```python
df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]})
print(df)
print('drop all duplicates values :- \n',df.drop_duplicates())
print('drop spacific columns duplicates values :- \n',df.drop_duplicates(subset='brand'))
print('drop spacific columns duplicates values :- \n',df.drop_duplicates(subset='brand'))
```

* To remove duplicates and keep last occurrences, use keep

```python
print('drop spacific columns duplicates values :- \n',df.drop_duplicates(subset=['brand', 'style']
                                                                         , keep='last'))
```
